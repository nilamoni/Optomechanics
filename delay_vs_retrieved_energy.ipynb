# Pulse delay vs retrieved energy

## Setup

The BEC is confined in a toroidal trap located at the center of an optical cavity. The potential experienced by each atom of mass $m$ in the condensate is given by:

$$
U(\rho,  z) = \frac{1}{2} m\omega_\rho(\rho - R)^2 + \frac{1}{2} m\omega_z z^2
$$

where, $\omega_\rho$ and $\omega_z$ are the harmonic trapping frequencies alon the radial and $z$ directions, $R$ is the radius of the ring trap. In the potential $U(\rho,z)$ the dynamics along the radial ($\rho$), axial ($z$), and azimuthal ($\phi$) directions decouple. We assume that all atoms remain in the same quantum state along the radial and axial directions during dynamical evolution; we focus instead on the azimuthal motion of the atoms, i.e., along $\phi$, which is not subject to any trapping.

A superposition of two frequency-degenerate optical beams derived from the same laser and carrying orbital angular momentum (OAM) $\pm l\hbar$ is now injected into the cavity to probe the BEC. Such coherent superpositions have been experimentally demonstrated to create an angular lattice inside the cavity about its axis. The beams are blue detuned far from the ground-to-excited state atomic transition and therefore interact weakly with the atoms via the dipole force, with the effect of spontaneous photon scattering being negligible.

## Hamiltonian

The motion of the BEC in the azimuthal ($\phi$) direction in presence of probe field with frequency, $\omega_p$ and control (drive) field with frequency,$\omega_l$ is described in the frame rotating at the laser drive frequency, by the one-dimensionall Hamiltonian:

$$
\hat{H} = \int^{2\pi}_{0}\hat{\psi}^\dagger(\phi)\left[-\frac{\hbar^2}{2I}\frac{d^2}{d\phi^2} + \hbar U_0\cos^2(l\phi)\hat{a}^\dagger\hat{a}\right]\hat{\Psi}(\phi)d\phi + \frac{g}{2}\int^{2\pi}_0\hat{\Psi}^\dagger(\phi)\hat{\Psi}^\dagger(\phi)\hat{\Psi}(\phi)\hat{\Psi}(\phi)d\phi - \hbar\Delta_0\hat{a}^\dagger\hat{a} + i\hbar\varepsilon_l\left(\hat{a}^\dagger - \hat{a}\right)  + i\hbar\varepsilon_p\left(a^\dagger e^{-i(\omega_p - \omega_l) t} -\hat{a}e^{i(\omega_p - \omega_l) t}\right) 
$$

Here, $\hat{a}^\dagger, \hat{a}$ are the creation and annhillation operators of the field inside the cavity, $\hat{\Psi}^\dagger, \hat{\Psi}$ are the Bosonic creation and annhillation operators of the atoms in the BEC such that, $[\hat{\Psi}(\phi),\Psi^\dagger(\phi^\prime)] = \delta(\phi - \phi^\prime)$ and the photonic operators obey, $[\hat{a},\hat{a}^\dagger] = 1$, $I = mR^2$ is the moment of inertia.the atomic moment of inertia about the cavity axis, $\Delta_0 = \omega_l - \omega_0$ is the detuning of the driving laser with respect to the cavity resonance frequency, $\omega_0$, $g = 2\hbar\omega_\rho a_{Na}/R$ is the two-body atomic interaction strength, $a_{Na}$ is the ground state scattering length of Sodium atoms in a condensate, $\rho$ is the radial coordinate,

#### Note: The detuning here is defined differently then what is commonly used which is $\Delta_0 = \omega_0 - \omega_l$

The presence of the optical lattice causes some atoms in the condensate to coherently Bragg scatter from their rotational state with winding number $L_p$ to states with $L_p \pm  2nl$, where $n =  1, 2, 3,\cdots ...$ The linear analog of such matter-wave scattering from an optical lattice inside a cavity has already been demonstrated.We assume the dipole potential to be weak (i.e., smaller than the chemical potential of the rotating condensate), and in that case the number of atoms scattered is small and only first order diffraction, $L_p\rightarrow L_p \pm 2l$, is appreciable. Based on this physical picture, a possible ansatz for the atomic field is:

$$
\hat{\Psi}(\phi) = \frac{e^{iL_p\phi}}{\sqrt{2\pi}}\hat{c}_p + \frac{e^{i(L_p+2l)\phi}}{\sqrt{2\pi}}\hat{c}_+ + \frac{e^{i(L_p - 2l)\phi}}{\sqrt{2\pi}}\hat{c}_-
$$

where, $\pm l$ is the winding number of the drive laser and the atomic operators obey, $[c_i,c^\dagger_j] = \delta_{i,j},\; (i,j) = p,+, -$. The first term in the above equation corresponds to the original persistent current and the remaining two terms are the side modes excited by matter wave diffraction.

$$
\hat{c}_p^\dagger \hat{c}_p + \hat{c}_+^\dagger \hat{c}_+ + \hat{c}_-^\dagger \hat{c}_- = \hat{N
}$$

with $\hat{N}$ being the number operator for the number of atoms in the BEC. Substituting the above two equations in the original Hamiltonian we get, the Hamiltonian of the system containing two fields, control (drive) and probe (signal) in a frame rotating with the frequency of the control field, $\omega_l$ is given by: 

$$
\begin{equation}
\hat{H} = \hbar\omega_c\hat{c}^\dagger\hat{c} + \hbar\omega_d\hat{d}^\dagger\hat{d} + \hbar\left[G(\hat{X}_c + \hat{X}_d) - \tilde{\Delta}\right]\hat{a}^\dagger\hat{a} + i\hbar\varepsilon_l\left(\hat{a}^\dagger - \hat{a}\right) + i\hbar\varepsilon_p\left(a^\dagger e^{-i(\omega_p - \omega_l) t} -\hat{a}e^{i(\omega_p - \omega_l) t}\right) + \hbar\tilde{g}\tilde{C},
\end{equation}
$$

where,

$$
\omega_c = \frac{\hbar(L_p + 2l)^2}{2I},\quad \text{and}\quad \omega_d = \frac{\hbar(L_p - 2l)^2}{2I}
$$

$$
\begin{align}
\hat{X}_{c(d)} &= \frac{\hat{c}^\dagger(\hat{d}^\dagger) + \hat{c}(\hat{d})}{\sqrt{2}}\\
%
\hat{c} &= \frac{c_p^*\hat{c}_+}{\sqrt{N}},\; \hat{d} = \frac{c_p^*\hat{c}_-}{\sqrt{N}},
\end{align}
$$

Here, we have assumed that the number of atoms in the side modes is small, and the mode with winding number $L_p$ is macroscopically occupied (i.e., its dynamics are classical), therefore, we posit $\hat{c}^\dagger_p\hat{c} \approx N$ and introduce the operators, $\hat{c}$ and $\hat{d}$. Note: $\hat{c}_p\equiv c_p$ is now simply a complex number and not an operator

$$
\begin{align}
\tilde{\Delta} &= \Delta_0 - U_0\frac{\hat{N}}{2},\quad\left(\Delta_0 = \omega_l - \omega_0\right)\\
%
U_0 &= \frac{g^2_a}{\Delta_a}
%
\end{align}
$$

Where, $\Delta_a$ is the detuning of the optical frequency from the atomic transition and $g_a$ is the strength of the interaction between one photon an one atom.

$$
\begin{align}
%
\tilde{g} &= \frac{g}{4\pi\hbar},\\
%
G &= U_0\frac{\sqrt{N}}{2\sqrt{2}},\\
%
\varepsilon_{p(l)} &= \sqrt{\frac{\mathcal{P}^{(p(l))}_{\text{in}}\gamma_0}{\hbar\omega_{p(l)}}}\quad \text{Input field amplitude}
\end{align}
$$

where, $\mathcal{P}^{(p(l))}$ is the input power of probe (p) and control (l), $\omega_{p(l)}$ is the frequency of the probe (p) and control (l), $\gamma_0$ is the cavity decay rate.

Finally, $\hbar\tilde{g}\tilde{C}$ in the Hamiltonian represents the effect of atomic interactions. 


$$
H_{\text{Int}} = 4\hbar\tilde{g}N\left(\hat{c}^\dagger\hat{c} + \hat{d}^\dagger\hat{d}\right) + 2\hbar\tilde{g}N\left(\hat{c}\hat{d} + \hat{c}^\dagger\hat{d}^\dagger\right)
$$

where,

$$
\tilde{g} = \frac{\omega_\rho a_{\text{Na}}}{2\pi R} = 6.96303\times1-^{-10}\gamma_0\; \text{for the considered parameterss}
$$

In derving the above Hamiltonian, we have ignored higher order terms of $\hat{X}_{c(d)}$.

Therefore, the total Hamiltonian is given by:

$$
\begin{equation}
\hat{H} = \hbar\omega_c\hat{c}^\dagger\hat{c} + \hbar\omega_d\hat{d}^\dagger\hat{d} + \hbar\left[G(\hat{X}_c + \hat{X}_d) - \tilde{\Delta}\right]\hat{a}^\dagger\hat{a} + i\hbar\varepsilon_l\left(\hat{a}^\dagger - \hat{a}\right) + i\hbar\varepsilon_p\left(a^\dagger e^{-i(\omega_p - \omega_l) t} -\hat{a}e^{i(\omega_p - \omega_l) t}\right) + 4\hbar\tilde{g}N\left(\hat{c}^\dagger\hat{c} + \hat{d}^\dagger\hat{d}\right) + 2\hbar\tilde{g}N\left(\hat{c}\hat{d} + \hat{c}^\dagger\hat{d}^\dagger\right)
\end{equation}
$$


## Heisenberg's equations of motion

$$
\partial_t\hat{x} = \frac{i}{\hbar}[\hat{H}, \hat{x}]
$$

Using relations, $[\hat{a}^\dagger\hat{a}, \hat{a}] = -\hat{a}$, $[\hat{a}, \hat{a}^\dagger] = 1\implies [\hat{a}^\dagger, \hat{a}] = -1$, and $[\hat{x}^\dagger\hat{x}, \hat{y}] =0$ and taking expectation on both sides we get:


$$
\begin{align}
\frac{d<\hat{a}>}{dt} &=  -i\left[G(<\hat{X}_c> + <\hat{X}_d>) - \tilde{\Delta}\right]<\hat{a}>+\varepsilon_l + \varepsilon_pe^{-i(\omega_p - \omega_l) t} -\frac{\gamma_0}{2}<\hat{a}>\\
%
\frac{d<\hat{c}>}{dt} &= -\left(\frac{\gamma_m}{2} + i\left(\omega_c + 4\tilde{g}N\right)\right)<\hat{c}> - i\frac{G}{\sqrt{2}}<\hat{a}^\dagger\hat{a}>  -2i\tilde{g}N<\hat{d}^\dagger>\\
%
\frac{d<\hat{d}>}{dt} &= -\left(\frac{\gamma_m}{2} + i\left(\omega_d + 4\tilde{g}N\right)\right)<\hat{d}> - i\frac{G}{\sqrt{2}}<\hat{a}^\dagger\hat{a}> -2i\tilde{g}N<\hat{c}^\dagger>\\
\end{align}
$$

where,

$$
<\hat{X}_{c(d)}> = \frac{<\hat{c}^\dagger>(<\hat{d}^\dagger>) + <\hat{c}>(<\hat{d}>)}{\sqrt{2}} = \frac{\beta_{c(d)} + \beta^*_{c(d)}}{\sqrt{2}}
$$

We now make a substitution,

$$
<\hat{a}> = \alpha + \alpha_+ e^{-i(\omega_p - \omega_l)t},\quad <\hat{c}(\hat{d})> = \beta_{c(d)} + \beta^+_{c(d)} e^{-i(\omega_p - \omega_l)t}
$$

in the above equations of motion to get and collecting the coefficient of $e^{-(\omega_p - \omega_l)t}$:

$$
\begin{align}
\frac{d\alpha}{dt} &=  -i\left[\frac{G}{\sqrt{2}}(\beta_c + \beta^*_c + \beta_d + \beta^*_d) - \tilde{\Delta} \right]\alpha + \varepsilon_l -\frac{\gamma_0}{2}\alpha\\
%
\frac{d\beta_{c}}{dt} &= -(\frac{\gamma_m}{2} + i\omega_c+ 4i\tilde{g}N)\beta_{c} - i\frac{G}{\sqrt{2}}|\alpha|^2\\
%
\frac{d\beta_{d}}{dt} &= -(\frac{\gamma_m}{2} + i\omega_d+ 4i\tilde{g}N)\beta_{d} - i\frac{G}{\sqrt{2}}|\alpha|^2
\end{align}
$$

The solution to above equation at steady state are:

$$
\alpha = \frac{\varepsilon_l}{i\left[\frac{G}{\sqrt{2}}(\beta_c + \beta^*_c + \beta_d + \beta^*_d) - \tilde{\Delta} \right] + \frac{\gamma_0}{2}}
$$

$$
\beta_{c(d)} =  \frac{-iG|\alpha|^2}{\sqrt{2}}\frac{1}{\frac{\gamma_m}{2} + i\omega_{c(d)}}
$$


and

$$
\begin{align}
\frac{d\alpha_+}{dt} &=  -\left(\frac{\gamma_0}{2} +i \tilde{\Delta}_p\right)\alpha_+ -i\tilde{G}(\beta^+_c + \beta^+_d) + \varepsilon_p -i\tilde{G}((\beta^+_c)^* + (\beta^+_d)^*)e^{2i(\omega_p - \omega_l)t} \\
%
\frac{d\beta^+_{c}}{dt} &= -\left(\frac{\gamma_m}{2} + i\left(\tilde{\Delta}_c+ 4\tilde{g}N\right)\right)\beta^+_{c} - i\tilde{G}\alpha_+ - 2i\tilde{g}N\beta^*_ce^{2i(\omega_p - \omega_l)t}\\
%
\frac{d\beta^+_{d}}{dt} &= -\left(\frac{\gamma_m}{2} + i\left(\tilde{\Delta}_d+ 4\tilde{g}N\right)\right)\beta^+_{d} - i\tilde{G}\alpha_+ -2i\tilde{g}N\beta^*_de^{2i(\omega_p - \omega_l)t}
\end{align}
$$

where, 

$$
\begin{align}
\tilde{\Delta}_{c(d)} &= \omega_{c(d)} - \omega_p + \omega_l\\
%
\tilde{\Delta}_p &= - \omega_p + \omega_l -\tilde{\Delta} = -\omega_p +\omega_l - (\Delta_0 - U_0\frac{N}{2}) = -\omega_p +\omega_l - (\omega_l - \omega_0) + U_0\frac{N}{2} = \omega_0 -\omega_p + U_0\frac{N}{2} =  \Delta_p + U_0\frac{N}{2}\; \left(\Delta_p = \omega_0 - \omega_p\right)\\
%
\tilde{G} &= \frac{G}{\sqrt{2}}\alpha(t)
\end{align}
$$

For storage and retrieval of the probe pulse, it is made resonant with the cavity, i.e., $\omega_p = \omega_0$. This implies, $\Delta_ p = \omega_0 - \omega_p = 0$ and thus, $\tilde{\Delta_p} = U_0\frac{N}{2}$

Again, the control or drive field is detuned at a frequency $\omega_c$ lesser than the cavity resonance frequency, i.e., $\omega_0 - \omega_l = \omega_c$. Therefore, $\tilde{\Delta}_c = \omega_c - \omega_p + \omega_l = \omega_c - \omega_0 + (\omega_0 -\omega_c) = 0$

And, $\tilde{\Delta}_d = \omega_d - \omega_p + \omega_l = \omega_d - \omega_0 + (\omega_0 -\omega_c) = \omega_d -\omega_c$.

#### Summaries everything:

$$
\begin{align}
\omega_p &= \omega_0\implies \Delta_p = \omega_0-\omega_p = 0\\
%
\omega_l &= \omega_0-\omega_c\\
%
\tilde{\Delta}_p &= \Delta_p + U_0\frac{N}{2} = U_0\frac{N}{2}\approx 0\\
%
\tilde{\Delta}_c &= 0\\
%
\tilde{\Delta}_d &= \omega_d - \omega_c
\end{align}
$$

The quantity $U_0\frac{N}{2}\approx 0$ for the chosen parameters.


Now lets say we keep $\omega_p = \omega_0$ and take $\omega_l = \omega_0 - \omega^\prime$ where $\omega^\prime$ is some random frequency, then

$$
\begin{align}
\omega_p &= \omega_0\implies \Delta_p = \omega_0-\omega_p = 0\\
%
\omega_l &= \omega_0-\omega^\prime\\
%
\tilde{\Delta}_p &= \Delta_p + U_0\frac{N}{2} = U_0\frac{N}{2}\approx 0\\
%
\tilde{\Delta}_c &= \omega_c - \omega_p + \omega_l = \omega_c - \omega_0 + (\omega_0 -\omega^\prime) = \omega_c - \omega^\prime\\
%
\tilde{\Delta}_d &= \omega_d - \omega_p + \omega_l = \omega_d - \omega_0 + (\omega_0 -\omega^\prime) = \omega_d - \omega^\prime
\end{align}
$$

If the difference between $\omega_c - \omega_d = \omega_{\text{diff}}$ then,

$$
\begin{align}
\tilde{\Delta}_c &= \omega_c - \omega^\prime\\
%
\tilde{\Delta}_d &= (\omega_c - \omega_{\text{diff}}) - \omega^\prime\implies \tilde{\Delta}_d = \tilde{\Delta}_c-\omega_{\text{diff}}
\end{align}
$$

# Note:

#### What is $<\hat{a}^\dagger \hat{a}>$?

Ans: Please remember that the general state $|\psi>$ for the composite system is $|n_c>\otimes|n_d>\otimes|n_a>\equiv |n_{c,d,a}>$. Therefore, 
$<\hat{a}^\dagger \hat{a}> = <n_{c,d,a}|\hat{a}^\dagger \hat{a}|n_{c,d,a}>$. Now, since our laser is a coherent source, $|n_a>$ is chosen to be a coherent state, $|\alpha>$ and not the number state $|n\rangle$. Therefore, $\hat{a}|\alpha>  = \alpha|\alpha>$ and $\hat{a}^\dagger|\alpha\rangle = \alpha^*|\alpha>$. Therefore, $<\hat{a}^\dagger \hat{a}> = |\alpha|^2$.

# Field definitions

```python
# Define the function for the time-dependent Gaussian function
def gaussian(t, center, width):
    return np.exp(-(t - center)**2/(2*width**2))

# Define other functions
def G_tilde(t, t_read):
    eta_w = 2*pi*0.48e6/gamma_00
    eta_r = 2*pi*0.79e6/gamma_00

    G_temp_1 = eta_w*gaussian(t, t_write, tau_l) + eta_r*gaussian(t, t_read, tau_l)

    return G_temp_1

def epsilon_p(t):

    return eta_p*gaussian(t, t_write, tau_p)
```

# Equations of motion

```python
# Define the coupled differential equations
def equations(t, y, delta_tilde_c, tau_read):
    
    # final detunings
    delta_tilde_p = 0
    omega_diff    = 3
    delta_tilde_d = delta_tilde_c - omega_diff
    
    dydt = np.zeros(3, dtype=np.complex128)
    
    dydt[0] = -(gamma_0/2 + iota*delta_tilde_p)*y[0] - iota*G_tilde(t, tau_read)*(y[1] + y[2]) + epsilon_p(t) #- iota*G_tilde(t)*(np.conj(y[1]) + np.conj(y[2]))*np.exp(2*iota*(omega_p - omega_l)*t)
    
    dydt[1] = -(gamma_m/2 + iota*(delta_tilde_c + 4*g_tilde*N))*y[1] - iota*G_tilde(t, tau_read)*y[0] #- 2*iota*g_tilde*N*np.exp(2*iota*(omega_p - omega_l)*t)*np.conj(y[1]) #- iota*G_tilde(t)*np.conj(y[0])*np.exp(2*iota*(omega_p - omega_l)*t)
    
    dydt[2] = -(gamma_m/2 + iota*(delta_tilde_d + 4*g_tilde*N))*y[2] - iota*G_tilde(t, tau_read)*y[0] #- 2*iota*g_tilde*N*np.exp(2*iota*(omega_p - omega_l)*t)*np.conj(y[2]) #- iota*G_tilde(t)*np.conj(y[0])*np.exp(2*iota*(omega_p - omega_l)*t)

    return dydt
```

```python
# define a vector to store the read times
tau_read = np.linspace(70, 900, int(np.abs(900 - 70)/2))

# Create a list to store the retrieved amplitude
retrieved_amplitude = np.zeros(np.size(tau_read))

# define a time vector
t_min = 0
t_max = 1000
t_span = (t_min, t_max)
t = np.linspace(t_span[0], t_span[1], int(np.abs(t_max - t_min)*10))


for index, delay in enumerate(tau_read):
        # Initial conditions
        y0 = [complex(0,0),complex(0,0), complex(0,0)]       # Initial values for y1, y2, and y3

        # Solve the differential equations
        sol = solve_ivp(equations, t_span, y0, args=(0,delay),  t_eval=np.linspace(t_span[0], t_span[1], int(np.abs(t_max - t_min)*10)))

        # store the field amplitude
        y1 = np.abs(sol.y[0])

        # Find first and second maxima
        arr = y1
        first_max = np.max(arr)
        
        # Find indices where the value is tau_read +- some small number
        indices = np.where((t >= delay - 10) & (t <= delay + 10))
        
        second_max = np.max(arr[indices])

        retrieved_amplitude[index] = second_max/first_max
```

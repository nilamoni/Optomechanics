# Fourier analysis of the intracavity field

## Setup

The BEC is confined in a toroidal trap located at the center of an optical cavity. The potential experienced by each atom of mass $m$ in the condensate is given by:

$$
U(\rho,  z) = \frac{1}{2} m\omega_\rho(\rho - R)^2 + \frac{1}{2} m\omega_z z^2
$$

where, $\omega_\rho$ and $\omega_z$ are the harmonic trapping frequencies alon the radial and $z$ directions, $R$ is the radius of the ring trap. In the potential $U(\rho,z)$ the dynamics along the radial ($\rho$), axial ($z$), and azimuthal ($\phi$) directions decouple. We assume that all atoms remain in the same quantum state along the radial and axial directions during dynamical evolution; we focus instead on the azimuthal motion of the atoms, i.e., along $\phi$, which is not subject to any trapping.

A superposition of two frequency-degenerate optical beams derived from the same laser and carrying orbital angular momentum (OAM) $\pm l\hbar$ is now injected into the cavity to probe the BEC. Such coherent superpositions have been experimentally demonstrated to create an angular lattice inside the cavity about its axis. The beams are blue detuned far from the ground-to-excited state atomic transition and therefore interact weakly with the atoms via the dipole force, with the effect of spontaneous photon scattering being negligible.

## Hamiltonian

The motion of the BEC in the azimuthal ($\phi$) direction in presence of probe field with frequency, $\omega_p$ and control (drive) field with frequency,$\omega_l$ is described in the frame rotating at the laser drive frequency, by the one-dimensionall Hamiltonian:

$$
\hat{H} = \int^{2\pi}_{0}\hat{\psi}^\dagger(\phi)\left[-\frac{\hbar^2}{2I}\frac{d^2}{d\phi^2} + \hbar U_0\cos^2(l\phi)\hat{a}^\dagger\hat{a}\right]\hat{\Psi}(\phi)d\phi + \frac{g}{2}\int^{2\pi}_0\hat{\Psi}^\dagger(\phi)\hat{\Psi}^\dagger(\phi)\hat{\Psi}(\phi)\hat{\Psi}(\phi)d\phi - \hbar\Delta_0\hat{a}^\dagger\hat{a} + i\hbar\eta_l\left(\hat{a}^\dagger - \hat{a}\right)  + i\hbar\eta_p\left(a^\dagger e^{-i(\omega_p - \omega_l) t} -\hat{a}e^{i(\omega_p - \omega_l) t}\right) 
$$

Here, $\hat{a}^\dagger, \hat{a}$ are the creation and annhillation operators of the field inside the cavity, $\hat{\Psi}^\dagger, \hat{\Psi}$ are the Bosonic creation and annhillation operators of the atoms in the BEC such that, $[\hat{\Psi}(\phi),\Psi^\dagger(\phi^\prime)] = \delta(\phi - \phi^\prime)$ and the photonic operators obey, $[\hat{a},\hat{a}^\dagger] = 1$, $I = mR^2$ is the moment of inertia.the atomic moment of inertia about the cavity axis, $\Delta_0 = \omega_l - \omega_0$ is the detuning of the driving laser with respect to the cavity resonance frequency, $\omega_0$, $g = 2\hbar\omega_\rho a_{Na}/R$ is the two-body atomic interaction strength, $a_{Na}$ is the ground state scattering length of Sodium atoms in a condensate, $\rho$ is the radial coordinate,


The presence of the optical lattice causes some atoms in the condensate to coherently Bragg scatter from their rotational state with winding number Lp to states with $L_p \pm  2nl$, where $n =  1, 2, 3,\cdots ...$ The linear analog of such matter-wave scattering from an optical lattice inside a cavity has already been demonstrated.We assume the dipole potential to be weak (i.e., smaller than the chemical potential of the rotating condensate), and in that case the number of atoms scattered is small and only first order diffraction, $L_p\rightarrow L_p \pm 2l$, is appreciable. Based on this physical picture, a possible ansatz for the atomic field is:

$$
\hat{\Psi}(\phi) = \frac{e^{iL_p\phi}}{\sqrt{2}\pi}\hat{c}_p + \frac{e^{i(L_p+2l)\phi}}{\sqrt{2}\pi}\hat{c}_+ + \frac{e^{i(L_p - 2l)\phi}}{\sqrt{2}\pi}\hat{c}_-
$$

where, $\pm l$ is the winding number of the drive laser and the atomic operators obey, $[c_i,c^\dagger_j] = \delta_{i,j},\; (i,j) = p,+, -$. The first term in the above equation corresponds to the original persistent current and the remaining two terms are the side modes excited by matter wave diffraction.

$$
\hat{c}_p^\dagger \hat{c} + \hat{c}_+^\dagger \hat{c}_+ + \hat{c}_-^\dagger \hat{c}_- = N
$$

with $N$ being the total number of atoms in the BEC. Substituting the above two equations in the original Hamiltonian we get, the Hamiltonian of the system containing two fields, control (drive) and probe (signal) in a frame rotating with the frequency of the control field, $\omega_l$ is given by: 

$$
\begin{equation}
\hat{H} = \hbar\omega_c\hat{c}^\dagger\hat{c} + \hbar\omega_d\hat{d}^\dagger\hat{d} + \hbar\left[G(\hat{X}_c + \hat{X}_d) - \tilde{\Delta}\right]\hat{a}^\dagger\hat{a} + i\hbar\eta_l\left(\hat{a}^\dagger - \hat{a}\right) + i\hbar\eta_p\left(a^\dagger e^{-i(\omega_p - \omega_l) t} -\hat{a}e^{i(\omega_p - \omega_l) t}\right) + \hbar\tilde{g}\tilde{C},
\end{equation}
$$

where,

$$
\omega_c = \frac{\hbar(L_p + 2l)^2}{2I},\quad \text{and}\quad \omega_d = \frac{\hbar(L_p - 2l)^2}{2I}
$$

$$
\begin{align}
\hat{X}_{c(d)} &= \frac{\hat{c}^\dagger(\hat{d}^\dagger) + \hat{c}(\hat{d})}{\sqrt{2}}\\
%
\hat{c} &= \frac{c_p^*\hat{c}_+}{\sqrt{N}},\; \hat{d} = \frac{c_p^*\hat{c}_-}{\sqrt{N}},
\end{align}
$$

Here, we have assumed that the number of atoms in the side modes is small, and the mode with winding number $L_p$ is macroscopically occupied (i.e., its dynamics are classical), therefore, we posit $\hat{c}^\dagger_p\hat{c} \approx N$ and introduce the operators, $\hat{c}$ and $\hat{d}$. Note: $\hat{c}_p$ is now simply a complex number and not an operator

$$
\begin{align}
\tilde{\Delta} &= \Delta_0 - U_0\frac{N}{2\sqrt{2}},\quad\left(\Delta_0 = \omega_l - \omega_0\right)\\
%
U_0 &= \frac{g^2_a}{\Delta_a}
%
\end{align}
$$

Where, $\Delta_a$ is the detning of the optical frequency from the atomic transition and $g_a$ is the strength of the interaction between one photon an one atom.

$$
\begin{align}
%
\tilde{g} &= \frac{g}{4\pi\hbar},\\
%
G &= U_0\frac{\sqrt{N}}{2\sqrt{2}},\\
%
\eta_{p(l)} &= \sqrt{\frac{\mathcal{P}^{(p(l))}_{\text{in}}\gamma_0}{\hbar\omega_{p(l)}}}\quad \text{Input field amplitude}
\end{align}
$$

where, $\mathcal{P}^{(p(l))}$ is the input power of probe (p) and control (l), $\omega_{p(l)}$ is the frequency of the probe (p) and control (l), $\gamma_0$ is the cavity decay rate.

Finally, $\hbar\tilde{g}\tilde{C}$ in the Hamiltonian represents the effect of atomic interactions. We assume it to be zero for now.




## Heisenberg's equations of motion

$$
\partial_t\hat{x} = \frac{i}{\hbar}[\hat{H}, \hat{x}]
$$

Using relations, $[\hat{a}^\dagger\hat{a}, \hat{a}] = -\hat{a}$, $[\hat{a}, \hat{a}^\dagger] = 1$ and $[\hat{x}^\dagger\hat{x}, \hat{y}] =0$ and taking expectation on both sides we get:


$$
\begin{align}
\frac{d<\hat{a}>}{dt} &=  -i\left[G(<\hat{X}_c> + <\hat{X}_d>) - \tilde{\Delta})\right]<\hat{a}>+\eta_l + \eta_pe^{-i(\omega_p - \omega_l) t} -\frac{\gamma_0}{2}<\hat{a}>\\
%
\frac{d<\hat{c}>}{dt} &= -(\frac{\gamma_m}{2} + i\omega_c)<\hat{c}> - i\frac{G}{\sqrt{2}}<\hat{a}^\dagger\hat{a}>\\
%
\frac{d<\hat{d}>}{dt} &= -(\frac{\gamma_m}{2} + i\omega_d)<\hat{d}> - i\frac{G}{\sqrt{2}}<\hat{a}^\dagger\hat{a}>
\end{align}
$$

where,

$$
<\hat{X}_{c(d)}> = \frac{<\hat{c}^\dagger>(<\hat{d}^\dagger>) + <\hat{c}>(<\hat{d}>)}{\sqrt{2}} = \frac{\beta_{c(d)} + \beta^*_{c(d)}}{\sqrt{2}}
$$

# All parameters (SI)

```python
######################################
#         Actual parameters
######################################
    
# Mass of atom
m = 23*1.66054e-27         # in Kg

# Radius of Toroidal trap
R = 12e-6                  # in meters

# Moment of inertia
I = m*R**2

# Number of atoms
N = 1e4

# Angular momemtum numbers
L_p = 1
l   = 10

# Cavity decay rate
gamma_00 = 2*pi*2e6
```

# Parameters normalized w.r.t $\gamma_0$

```python
######################################
#         Normalized parameters
######################################

# Cavity decay rate
gamma_0 = 1

# Damping rate of each condensate
gamma_m = 2*pi*0.8/gamma_00

# Cavity resonance frequency
omega_0 = 2*pi*1e15/gamma_00

# Harmonic trapping frequencies
omega_rho = 2*pi*42/gamma_00
omega_z   = omega_rho/gamma_00

# Other parameters
G = 2*pi*7.5e3/gamma_00

# Matter wave frequencies
omega_c = (hbar*(L_p + 2*l)**2/(2*I))/gamma_00
omega_d = np.abs((hbar*(L_p - 2*l)**2/(2*I))/gamma_00)

# U
U = G*np.sqrt(2*N)/gamma_00

# drive detuning w.r.t cavity
delta_0 = -omega_c       # omega_l - omega_0

# effective detuning, Delta_tilde
delta_tilde = (delta_0 - U)

# power
power_l = 12.4e-15 # Femto watt
power_p = 12.4e-15*1e-2


# Probe frequency
omega_p = omega_0                # probe frequency is same as cavity resonance frequency
omega_l = (omega_0 - omega_c)    # control or drive is detuned to one matter wave frequency below cavity resonance frequency


# Field amplitudes
eta_l = np.sqrt(gamma_0*power_l/(hbar*omega_l))
eta_p = np.sqrt(gamma_0*power_p/(hbar*omega_p))
```

# Parameters normalized w.r.t $\omega_c$

```python
######################################
#         Normalized parameters
######################################

# Matter wave frequencies
omega_c0 = (hbar*(L_p + 2*l)**2/(2*I))
omega_d0 = np.abs(hbar*(L_p - 2*l)**2/(2*I))


omega_c = 1
omega_d = omega_d0/omega_c0

# Cavity decay rate
gamma_0 = gamma_00/omega_c0

# Damping rate of each condensate
gamma_m = 2*pi*0.8/omega_c0

# Cavity resonance frequency
omega_0 = 2*pi*1e15/omega_c0

# Harmonic trapping frequencies
omega_rho = 2*pi*42/omega_c0
omega_z   = omega_rho/omega_c0

# Other parameters
G = 2*pi*7.5e3/omega_c0

# U
U = G*np.sqrt(2*N)/omega_c0

# drive detuning w.r.t cavity
delta_0 = -omega_c       # omega_l - omega_0

# effective detuning, Delta_tilde
delta_tilde = (delta_0 - U)

# power
power_l = 12.4e-15 # Femto watt
power_p = 12.4e-15*1e-2


# Probe frequency
omega_p = omega_0                # probe frequency is same as cavity resonance frequency
omega_l = (omega_0 - omega_c)    # control or drive is detuned to one matter wave frequency below cavity resonance frequency


# Field amplitudes
eta_l = np.sqrt(gamma_0*power_l/(hbar*omega_l))
eta_p = np.sqrt(gamma_0*power_p/(hbar*omega_p))

```

# Equations of motion

```python

# Define the coupled differential equations
def equations(t, y):
    dydt = np.zeros(6, dtype=np.complex128)
    
    dydt[0] = -(gamma_0/2 - iota*delta_tilde)*y[0] - iota*G*(y[1]  + y[2] + y[3] + y[4] )*y[0]/np.sqrt(2) + eta_l + eta_p*np.exp(-iota*(omega_p - omega_l)*t)
    
    dydt[1] = -(gamma_0/2 + iota*delta_tilde)*y[1] + iota*G*(y[1]  + y[2] + y[3] + y[4] )*y[1]/np.sqrt(2) + eta_l + eta_p*np.exp(iota*(omega_p - omega_l)*t)
    
    dydt[2] = -(gamma_m/2 + iota*omega_c)*y[1] - iota*G*np.abs(y[0])**2/np.sqrt(2)
    
    dydt[3] = -(gamma_m/2 + iota*omega_d)*y[2] - iota*G*np.abs(y[0])**2/np.sqrt(2)
    
    dydt[4] = -(gamma_m/2 - iota*omega_c)*y[3] + iota*G*np.abs(y[0])**2/np.sqrt(2)
    
    dydt[5] = -(gamma_m/2 - iota*omega_d)*y[4] + iota*G*np.abs(y[0])**2/np.sqrt(2)

    return dydt
```

# Solve the equations of motion

```python
# Initial conditions
#y0 = [complex(0,0),complex(0,0), complex(0,0)]       # Initial values for y1, y2, and y3

y0 = [complex(0,0),complex(0,0), complex(0,0), complex(0,0), complex(0,0),complex(0,0)]       # Initial values for y1, y2, and y3

t_span = (0, 200)                                     # Time span for the solution

# Solve the differential equations
sol = solve_ivp(equations, t_span, y0, t_eval=np.linspace(t_span[0], t_span[1], int(200e2)))
```

# Fourier Transform of $<\hat{a}(t)> = \alpha(t)$, $<\hat{c}(t)> = \beta_c(t)$, and $<\hat{d}(t)> = \beta_d(t)$

```python
# Calculate the Fourier transform
fft_alpha = np.fft.fft(sol.y[0])
fft_beta_c = np.fft.fft(sol.y[1])
fft_beta_d = np.fft.fft(sol.y[2])

# define a time vector
t = np.linspace(t_span[0], t_span[1], int(200e2))

# Calculate the frequencies corresponding to the Fourier transform
freqs = np.fft.fftfreq(len(t), t[1] - t[0])
```
